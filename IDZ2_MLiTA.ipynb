### 1) Примеры и контрпримеры 

Примеры

    ([][])[()]  
    []()()

Контрпримеры
    
    ()[([])]  
    )()[]  
    ([](()))

### 2) КС-грамматика

/\ - пустой символ  
<A\> ::= (<B\>)<A\> | \[<B\>\]<A\> | /\\  
<B\> ::= ()<B\> | \[\]<B\> | /\\

### 4) Проверка однозначности ветвления по первому символу

Ветвление <A\> ::= (<B\>)<A\> | \[<B\>\]<A\> | /\\  
w1 - (<B\>)<A\>  
w2 - \[<B\>\]<A\>  
L(w1) = { ( }  
L(w2) = { \[ }  
L(w1) ∩ L(w2) = ∅

Ветвление <B\> ::= ()<B\> | \[\]<B\> | /\\  
w1 - ()<B\>  
w2 - \[\]<B\>  
L(w1) = { ( }  
L(w2) = { \[ }  
L(w1) ∩ L(w2) = ∅

### 8) Код на языке python

```python
class Parser:
    def __init__(self, seq: str):
        self.seq = seq
    
    def parse(self):
        return self.A(self.seq)
    
    def A(self, seq: str):
        if len(seq)!=0:
            if seq[0] == '(':
                seq = self.B(seq[1:])
                if seq:
                    if seq[0] == ')':
                        return self.A(seq[1:])
                    else: return False
                else: return False
            elif seq[0] == '[':
                seq = self.B(seq[1:])
                if seq:
                    if seq[0] == ']':
                        return self.A(seq[1:])
                    else: return False
                else: return False
            else: return False
        else: return True
                
    def B(self, seq: str):
        if len(seq)!=0:
            if seq[0] == '(':
                if seq[1] == ')':
                    seq = seq[2:]
                    return self.B(seq)
                else: return False
            elif seq[0] == '[':
                if seq[1] == ']':
                    seq = seq[2:]
                    return self.B(seq)
                else: return False
            else: return seq
        else: return False

seq = input('Enter the sequence: ')
parser = Parser(seq)
if parser.parse():
    print('Right\n')
else:
    print('Wrong\n')

seq = input('Enter the sequence: ')
parser = Parser(seq)
if parser.parse():
    print('Right')
else:
    print('Wrong')
```
